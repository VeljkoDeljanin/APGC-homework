```python
import numpy as np
from numpy import linalg as LA
```

```python
def two_equations(Ah, A1h):
    O3 = np.zeros(3)
    row1 = np.concatenate((O3, -A1h[2]*Ah, A1h[1]*Ah))
    row2 = np.concatenate((A1h[2]*Ah, O3, -A1h[0]*Ah))
    return np.array([row1, row2])
```

```python
def DLT(orig_pts: list, img_pts: list):
    A = []
    for i in range(len(orig_pts)):
        rows = two_equations(orig_pts[i], img_pts[i])
        A.append(rows[0])
        A.append(rows[1])

    A = np.array(A)
    U, S, Vh = LA.svd(A)

    P = Vh[-1].reshape(3, 3)
    P = P/P[2, 2]

    return P
```
